```python
import pandas as pd
from pathlib import Path

# Caminho do arquivo JSON de entrada
json_path = Path(r"C:\Users\Willgnner\Documents\trabalho-final-pmd\dados\dados_processos_cejusc_14052025.json")

# Caminho do arquivo Parquet de saída (mesmo nome, outra extensão)
parquet_path = json_path.with_suffix(".parquet")

print(f"Lendo JSON: {json_path}")

# Tenta ler como JSON em linhas (NDJSON); se falhar, tenta o formato padrão
try:
    df = pd.read_json(json_path, lines=True)
except ValueError:
    df = pd.read_json(json_path)

print(f"DataFrame carregado com {len(df)} linhas e {len(df.columns)} colunas.")
print(f"Salvando em Parquet: {parquet_path}")

df.to_parquet(parquet_path, index=False)

print("Conversão concluída com sucesso!")
```
