**Questão 1. Aquecimento: Definindo Funções Simples**

```python
# 1: Função bird_pad
def bird_pad(s):
    # Vai adicionar 'bird' antes e depois da string
    return "bird" + s + "bird"

print(bird_pad("goat"))  # Vai rodar: birdgoatbird
print(bird_pad("_"))     # Vai rodar: bird_bird
```

```python
# 2: Função print_n
def print_n(s, n):
    # Vai imprimir a string s n vezes, uma por linha
    for _ in range(n):
        print(s)

print_n("cat", 3)  # Saída do código
```

**Questão 2. Algoritmo de Euclides**

```python
# 1: Implementação do Algoritmo de Euclides
def gcd(a, b):
    # Aqui, vai retornar o MDC de dois números usando o Algoritmo de Euclides
    while b != 0:
        a, b = b, a % b
    return a

# Aqui, se quiser testar a função com base no MDC
print(gcd(48, 18))
```

```python
# 2: Saídas com pares de números
pares = [(1200, 300), (5040, 60), (29, 31), (2023, 2024)]

for x, y in pares:
    print(f"MDC({x}, {y}) = {gcd(x, y)}")  # Vai calcular e exibe o MDC de cada par
```

```python
# 3: Testando com números negativos
teste_negativo = (-48, -18)
resultado_negativo = gcd(*teste_negativo)
print(f"MDC{teste_negativo} = {resultado_negativo}")

# Explicação:
# O Algoritmo de Euclides funciona para números negativos da mesma forma que para positivos
# pois ele usa a operação de módulo (%) e troca os valores iterativamente até b ser zero
# Assim, o MDC(-a, -b) é igual ao MDC(a, b), já que o sinal não interfere no resultado
```

**Questão 3. Aproximando o Número de Euler e**

```python
# 1: Aproximando o Número de Euler

def euler_limit(n):
    # Vai calcular uma aproximação de e usando a fórmula de Bernoulli
    return (1 + 1/n) ** n  # Aplicando diretor na fórmula

# Saída
valores = [1, 10, 100, 1000, 10000]  # Estou usando diferente valores de n para testar a aproximação
for n in valores:
    print(f"Para n = {n}, e ≈ {euler_limit(n)}")

# A função usa a fórmula (1 + 1/n)ⁿ para calcular uma aproximação de e
# Quanto maior o valor de n, mais próximo o resultado estará de e ≈ 2.718
# O laço testa a função para diferentes valores de n, mostrando como a aproximação melhora
```

```python
# 2: Implementação do fatorial recursivo

def factorial(k):
    # Vai calcular o fatorial de k usando recursão.
    if k == 0 or k == 1:
        return 1  # Caso base: fatorial de 0 e 1 é 1
    return k * factorial(k - 1)  # Aplicando chamada recursiva

# Saída
valores = [0, 1, 5, 7]
for v in valores:
    print(f"{v}! = {factorial(v)}")  # Vai exibir o fatorial de cada número testado

# A função usa recursão para calcular o fatorial
# O caso base (k == 0 ou k == 1) retorna 1
# A chamada recursiva k * factorial(k - 1) calcula o fatorial para valores maiores
```

```python
# 3: Aproximação de e usando a soma infinita

def euler_infinite_sum(n):
    # Vai aproximando e somando os primeiros n termos da série infinita
    soma = 0
    for k in range(n):
        soma += 1 / factorial(k)  # Está usando a função da questão 2
    return soma

# Saídas do code
valores = [1, 5, 10, 15]  # Usando diferentes números de termos para a soma
for v in valores:
    print(f"Aproximação de e com {v} termos: {euler_infinite_sum(v)}")

# Usando a função factorial da Questão 2 para calcular cada termo
# Quanto maior o valor de n, mais próximo o resultado estará de e ≈ 2.718
```

```python
# 4: Encontrar o menor número de termos necessários para alcançar uma precisão t

def euler_approx(t):
    # Vai encontrar o menor número de termos necessários para que a aproximação de e esteja dentro do erro t
    verdadeiro_e = 2.718281828459045  # Valor real de e
    soma = 0
    k = 0

    while abs(soma - verdadeiro_e) > t:  # Continua somando até atingir a precisão desejada
        soma += 1 / factorial(k)  # Usando a função factorial da Questão 2
        k += 1

    return k  # Vai retorna o número mínimo de termos necessários

# Testes
tolerancias = [0.1, 0.01, 0.001, 0.0001]  # Testei com diferentes níveis de precisão
for t in tolerancias:
    print(f"Para erro {t}, são necessários {euler_approx(t)} termos.")

# A função calcula sucessivas aproximações de e somando os termos da série infinita
# O laço while continua somando termos até que o erro seja menor que t
# Quanto menor for t, mais termos serão necessários para atingir a precisão desejada
```

```python
# 5: Tabelas de aproximação do número de Euler

def print_euler_sum_table(n):
    # Vai imprimir os valores sucessivos da aproximação de e usando a soma infinita
    print("\nAproximação de e usando a soma infinita:")
    print("Termos\tValor Aproximado")
    for i in range(1, n + 1):
        print(f"{i}\t{euler_infinite_sum(i)}")  # Usa a função da Questão 3

def print_euler_lim_table(n):
    # Vai imprimir os valores sucessivos da aproximação de e usando a fórmula de Bernoulli
    print("\nAproximação de e usando a fórmula de Bernoulli:")
    print("n\tValor Aproximado")
    for i in range(1, n + 1):
        print(f"{i}\t{euler_limit(i)}")  # Usa a função da Questão 1

# Usando para exibir as tabelas
n = 10  # Usei o númer de termos a serem testados, poderia ser outro número
print_euler_sum_table(n)
print_euler_lim_table(n)

# print_euler_sum_table(n) mostra a aproximação de e usando a soma infinita da série
# print_euler_lim_table(n) mostra a aproximação de e usando a fórmula de Bernoulli
# Os valores aproximados são impressos em formato de tabela para melhor visualização
```

```python
# 6

# A aproximação de e pela série infinita (euler_infinite_sum) converge mais rápido do que pela fórmula de Bernoulli (euler_limit)
# Com poucos termos, a série infinita já se aproxima bastante do valor real de e, enquanto a fórmula de Bernoulli precisa de valores muito altos de n para alcançar a mesma precisão
# Então podemos falar que a série infinita é mais eficiente para calcular e com precisão rapidamente
```

**Questão 4. Testando Propriedades de um Inteiro**

```python
# 4: Testando se um número é potência de outro

def is_power_of_2(n):
    # Vai verificar se n é uma potência de 2
    if n < 1:
        return False
    return (n & (n - 1)) == 0  # Usa operação bitwise para verificar potência de 2

def is_power(b, n):
    # Vai verificar se n é uma potência de b (n = b^p para algum p)
    if n < 1 or b < 2:
        return False
    while n % b == 0:  # Vai dividir n por b enquanto for divisível
        n //= b
    return n == 1

# Usando valores para rodar o code
valores = [1, 2, 4, 8, 16, 18, 32, 64, 100]
for v in valores:
    print(f"{v} é potência de 2? {is_power_of_2(v)}")

# Rodando a função is_power
bases = [(3, 27), (5, 25), (2, 32), (4, 64), (6, 36)]
for base, num in bases:
    print(f"{num} é potência de {base}? {is_power(base, num)}")

# is_power_of_2(n) usa uma operação bitwise (n & (n - 1) == 0), que verifica se n tem apenas um bit ligado, o que indica que é uma potência de 2
# is_power(b, n) verifica se n é potência de b dividindo sucessivamente por b até chegar a 1. Se o resultado final for 1, então n = b^p
```
