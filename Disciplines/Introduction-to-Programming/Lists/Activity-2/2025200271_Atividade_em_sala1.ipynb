**Questão 1. Diversão com Strings**

```python
# 1(a)

def is_palindrome(s):
    # Vai verificar se a string é um palíndromo (sem ignorar espaços e capitalização)
    s = ''.join(s.lower().split())  # Converte para minúsculas e remove espaços
    return s == s[::-1]  # Vai comparar a string com sua versão invertida

# Saída do code
testes_a = ["level", "kayak", "pop", "hello", "Tacocat", "T A C O cat"]
for palavra in testes_a:
    print(f"'{palavra}' é palíndromo? {is_palindrome(palavra)}")

# Apenas verifica se a string é igual à sua versão invertida, sem alterar espaços ou letras maiúsculas
```

```python
# 1(b)

def is_palindrome_ignore(s):
    # Vai verificar  se a string é um palíndromo ignorando espaços e capitalização."""
    s = s.lower().replace(" ", "")  # Vai converter para minúsculas e remove espaços
    return s == s[::-1]  # Vai comparar a string com sua versão invertida

# Saída do code
testes_b = ["tacocat", "T A C O cat", "Was it a car or a cat I saw", "hello", "No lemon, no melon"]
for frase in testes_b:
    print(f"'{frase}' é palíndromo (ignorando espaços e letras maiúsculas)? {is_palindrome_ignore(frase)}")

# Vai ignorar espaços e capitalização antes da verificação, para que frases sejam tratadas corretamente
```

```python
# 2(a)

def is_abecedarian(s):
    # Vai verificar se a string está em ordem alfabética, ignorando espaços e capitalização
    s = s.lower().replace(" ", "")  # Vai converter para minúsculas e remove espaços

    # Vai percorrer a string verificando se cada letra está na ordem correta
    for i in range(len(s) - 1):
        if s[i] > s[i + 1]:  # Se uma letra for maior que a próxima, não é abecedária
            return False
    return True  # Se passou por todas as letras, é abecedária

# Saidá com algumas palavras
testes_a = ["adder", "beet", "dog", "cat", "floppy", "ghost"]
for palavra in testes_a:
    print(f"'{palavra}' é abecedária? {is_abecedarian(palavra)}")

# is_abecedarian(s)
# Vai converter a string para minúsculas e remove espaços
# Vai percorrer a string e verifica se cada letra vem antes ou é igual à próxima
# Se encontrar uma inversão, retorna False, caso contrário, retorna True
```

```python
# Questão 2(b)

def remove_vowels(s):
    # Vai remover vogais (a, e, i, o, u) da string sem usar str.replace
    vogais = "aeiouAEIOU"
    resultado = ""  # String onde vai armazenar a nova palavra sem vogais

    for letra in s:
        if letra not in vogais:  # Vai adicionar apenas consoantes e símbolos
            resultado += letra
    return resultado

# Saída com palavras
testes_b = ["cat?", "goAT!", "Hello World!", "Python"]
for palavra in testes_b:
    print(f"'{palavra}' sem vogais -> '{remove_vowels(palavra)}'")

# remove_vowels(s)
# Vai definir as vogais (a, e, i, o, u maiúsculas e minúsculas)
# ai criar uma nova string, adicionando apenas letras que não sejam vogais
# Não vai usar str.replace(), como pedido no enunciado
```

**Questão 2. Diversão com Listas**

```python
# 1

def list_reverse(lst):
    # Vai retorna a lista em ordem reversa. Se a entrada não for uma lista, exibe um erro
    if not isinstance(lst, list):  # Vai verificar se a entrada é uma lista
        return "Erro: A entrada precisa ser uma lista."
    return lst[::-1]  # Vai retornar a lista invertida

# Testes com númernos e palavras
testes = [
    [1, 2, 3],         # Lista comum;
    ["a", "b", "c"],   # Lista de strings;
    [10],              # Lista com um único elemento;
    [],                # Lista vazia;
    "não é uma lista", # Entrada inválida (string);
    12345              # Entrada inválida (número).
]

for teste in testes:
    print(f"Entrada: {teste} -> Saída: {list_reverse(teste)}")

# Vai verificar se a entrada é uma lista usando isinstance(lst, list). Se não for, retorna um erro
# Vai usar [::-1] para inverter a lista. Esse método é simples e eficiente
# Vai tratar diferentes casos, incluindo listas vazias, listas de um único elemento e entradas inválidas
```

```python
# 2

def is_sorted(seq):
    # Vai rertornar True se a sequência estiver em ordem não decrescente, False caso contrário
    for i in range(len(seq) - 1):  # Vai percorrer a lista até o penúltimo elemento
        if seq[i] > seq[i + 1]:  # Se um elemento for maior que o próximo, não está ordenado
            return False
    return True  # Se passou por todos sem problemas, a lista está ordenada

# Testes com números
testes = [
    [1, 2, 3, 4, 5],  # Ordenado;
    [3, 3, 4, 5, 5],  # Ordenado (valores repetidos são permitidos);
    [10],             # Lista com um único elemento (considerada ordenada);
    [],               # Lista vazia (considerada ordenada);
    [1, 2, 3, 2, 5],  # Não ordenado;
    [5, 4, 3, 2, 1],  # Não ordenado (decrescente).
]

# Executando
for seq in testes:
    print(f"Sequência: {seq} -> Está ordenada? {is_sorted(seq)}")

# Vai percorrer a lista apenas uma vez (O(n)) para verificar a ordenação
# Vai comparar elementos consecutivos, retornando False se algum valor estiver fora de ordem
# Se as listas vazias e de um único elemento são consideradas ordenadas, pois não há elementos fora de ordem
```

```python
# 3

def binary_search(lst, target):
    # Vai retornar True se target estiver na lista ordenada lst, False caso contrário
    left, right = 0, len(lst) - 1  # Define os limites da busca

    while left <= right:  # Enquanto houver elementos para buscar
        mid = (left + right) // 2  # Encontra o meio da lista

        if lst[mid] == target:  # Se encontrou o elemento, retorna True
            return True
        elif lst[mid] < target:  # Se o elemento está à direita, ajusta o limite esquerdo
            left = mid + 1
        else:  # Se o elemento está à esquerda, ajusta o limite direito
            right = mid - 1

    return False  # Se saiu do loop, o elemento não está na lista

# Testes com vários números
testes = [
    ([1, 2, 3, 4, 5], 3),  # Presente na lista;
    ([1, 2, 3, 4, 5], 6),  # Não está na lista;
    ([10, 20, 30, 40, 50], 50),  # Último elemento;
    ([10, 20, 30, 40, 50], 10),  # Primeiro elemento;
    ([7], 7),  # Lista com um único elemento, presente;
    ([7], 5),  # Lista com um único elemento, ausente;
    ([], 3)  # Lista vazia.
]

# Executando
for lista, valor in testes:
    print(f"Buscar {valor} em {lista} -> {binary_search(lista, valor)}")

# Vai Uusar busca binária para procurar o elemento de forma eficiente (O(log n)) (Visto na matéria de EDPA do mestrado)
# Vai dividir a lista ao meio a cada iteração, reduzindo o espaço de busca
# Vai considerar casos base, como lista vazia e lista com um único elemento
```

**Questão 3. Mais diversão com Strings**

```python
# 1

def char_hist(s):
    # Vai retorna um dicionário com a contagem de cada caractere da string, ignorando maiúsculas/minúsculas
    s = s.lower()  # Vai converter tudo para minúsculas para tratar 'G' e 'g' como o mesmo caractere
    histograma = {}  # Tenho aqui um dicionário para armazenar as contagens

    for char in s:
        if char in histograma:
            histograma[char] += 1  # Se já existe no dicionário, soma 1
        else:
            histograma[char] = 1  # Se não existe, inicia com 1

    return histograma

# Executando o code
testes_1 = ["Hello", "Banana", "GgGgG"]
for texto in testes_1:
    print(f"Histograma de '{texto}': {char_hist(texto)}")

# Vai converter a string para minúsculas (s.lower()) para tratar maiúsculas/minúsculas como iguais
# Vai converter cada caractere e conta quantas vezes aparece na string
# Vai armazenar os valores em um dicionário, onde a chave é o caractere e o valor é a quantidade de ocorrências
```

```python
# 2

def bigram_hist(s):
    # Vai retornar um dicionário com a contagem de bigramas (pares de caracteres consecutivos) na string
    s = s.lower()  # Vai converter para minúsculas para ignorar diferenças de maiúsculas/minúsculas
    histograma = {}  # Aqui tenhoo um dicionário para armazenar os bigramas e suas contagens

    for i in range(len(s) - 1):  # Vai percorrer a string até o penúltimo caractere
        bigrama = s[i] + s[i + 1]  # Forma o bigrama com dois caracteres consecutivos
        if bigrama in histograma:
            histograma[bigrama] += 1  # Se já existe, soma 1
        else:
            histograma[bigrama] = 1  # Se não existe, inicia com 1

    return histograma

# Testes para bigram_hist
testes_2 = ["Hello", "Banana", "Test"]
for texto in testes_2:
    print(f"Histograma de bigramas de '{texto}': {bigram_hist(texto)}")

# Vai transforma a string em minúsculas para que "He" e "he" sejam tratados como o mesmo bigrama
# Vai percorrer a string pegando pares consecutivos de caracteres
# Vai armazenar os bigramas em um dicionário, contando quantas vezes aparecem
```

**Questão 4. Tuplas como Vetores**

```python
# 1

def vec_scalar_mult(vec, scalar):
    # Vai multiplicar cada elemento de uma tupla por um número escalar
    if not isinstance(vec, tuple) or not isinstance(scalar, (int, float)):
        raise TypeError("O primeiro argumento deve ser uma tupla e o segundo um número.")

    return tuple(x * scalar for x in vec)

# Rodando para para vec_scalar_mult
testes_1 = [
    ((1, 2, 3), 2),
    ((-1, 0, 5), 3.5),
    ((4, 8, 12), 0)
]

# Saída do code
for v, s in testes_1:
    print(f"{v} * {s} -> {vec_scalar_mult(v, s)}")
```

```python
# 2

def vec_inner_product(vec1, vec2):
    # Vai calcular o produto interno de dois vetores representados por tuplas
    if not isinstance(vec1, tuple) or not isinstance(vec2, tuple):
        raise TypeError("Os argumentos devem ser tuplas.")
    if len(vec1) != len(vec2):
        raise ValueError("Os vetores devem ter o mesmo tamanho.")

    return sum(x * y for x, y in zip(vec1, vec2))

# Rodando para para vec_inner_product
testes_2 = [
    ((1, 2, 3), (4, 5, 6)),
    ((-1, 0, 5), (3, -2, 1)),
    ((2, 3), (4, 5))
]

# Saída do code
for v1, v2 in testes_2:
    print(f"Produto interno {v1} . {v2} -> {vec_inner_product(v1, v2)}")
```

```python
# 3

def check_valid_mx(matrix):
    # Vai verificar se o argumento é uma matriz válida representada por uma tupla de tuplas
    if not isinstance(matrix, tuple) or not all(isinstance(row, tuple) for row in matrix):
        return False
    if len(set(len(row) for row in matrix)) > 1:
        return False  # Vai verificar se todas as linhas têm o mesmo tamanho
    return True

# Rodando para check_valid_mx
testes_3 = [
    ((1, 2), (3, 4)),   # Matriz válida;
    ((1, 2, 3), (4, 5, 6)),  # Matriz válida;
    ((1, 2), [3, 4]),   # Contém lista em vez de tupla;
    ((1, 2, 3), (4, 5)),  # Linhas com tamanhos diferentes.
]

# Saída
for matriz in testes_3:
    print(f"Matriz {matriz} é válida? {check_valid_mx(matriz)}")
```

```python
# 4

def mx_vec_mult(matrix, vector):
    # Vai multiplicar uma matriz por um vetor, ambos representados por tuplas
    if not check_valid_mx(matrix):
        raise ValueError("A matriz fornecida não é válida.")
    if not isinstance(vector, tuple) or len(vector) != len(matrix[0]):
        raise ValueError("O vetor deve ser uma tupla com o mesmo número de colunas da matriz.")

    return tuple(sum(m * v for m, v in zip(row, vector)) for row in matrix)

# Rodando para mx_vec_mult
testes_4 = [
    (((1, 2), (3, 4)), (1, 1)),  # Multiplicação válida;
    (((2, 3, 4), (5, 6, 7)), (1, 2, 3)),  # Outra multiplicação válida.
]

for mat, vec in testes_4:
    print(f"{mat} * {vec} -> {mx_vec_mult(mat, vec)}")
```

vec_scalar_mult(vec, scalar), vai multiplicar cada elemento de uma tupla por um escalar
vec_inner_product(vec1, vec2), vai calcular o produto interno entre dois vetores (tuplas)
check_valid_mx(matrix), vai verificar se uma matriz é válida (todas as linhas do mesmo tamanho e representadas como tuplas)
mx_vec_mult(matrix, vector), vai multiplica uma matriz por um vetor, garantindo compatibilidade

Questão 5. Mais diversão com Vetores


```python
# 1

def is_valid_sparse_vector(vec):
    # Vai verificar se um dicionário representa um vetor esparso válido
    # As chaves devem ser inteiros não negativos e os valores devem ser numéricos

    if not isinstance(vec, dict):  # O vetor deve ser um dicionário
        return False

    for key, value in vec.items():
        if not isinstance(key, int) or key < 0:  # As chaves devem ser inteiros não negativos
            return False
        if not isinstance(value, (int, float)):  # Os valores devem ser números
            return False

    return True  # Se passou por todas as verificações, é válido

# Rodando para is_valid_sparse_vector
testes_1 = [
    {0: 1.5, 2: 3, 5: -2},  # Válido;
    {1: 0, 4: 2.3, 7: 5},  # Válido;
    {"a": 1, 2: 3},  # Chave não é número;
    {0: 1, -2: 3},  # Chave negativa;
    {0: "x", 2: 3},  # Valor não numérico.
]

# Executando
for vec in testes_1:
    print(f"Vetor {vec} é válido? {is_valid_sparse_vector(vec)}")
```

```python
# 2

def sparse_inner_product(vec1, vec2):
    # Vai calcular o produto interno de dois vetores esparsos representados por dicionários

    if not is_valid_sparse_vector(vec1) or not is_valid_sparse_vector(vec2):
        raise ValueError("Ambos os vetores devem ser vetores esparsos válidos.")

    produto = sum(vec1[k] * vec2[k] for k in vec1 if k in vec2)  # Aqui, multiplica apenas índices comuns
    return produto

# Rodando para sparse_inner_product
testes_2 = [
    ({0: 1, 2: 3}, {0: 2, 2: 4}),  # 1*2 + 3*4 = 14
    ({1: 2, 3: 4}, {1: 5, 3: 6}),  # 2*5 + 4*6 = 34
    ({0: 1, 2: 3}, {1: 2, 3: 4}),  # Nenhum índice em comum -> 0
]

# Rodando
for v1, v2 in testes_2:
    print(f"Produto interno {v1} . {v2} -> {sparse_inner_product(v1, v2)}")
```
